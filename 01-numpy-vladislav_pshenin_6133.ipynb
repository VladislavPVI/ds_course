## II. Сгенерировать OLAP-куб, где фиксируем продажи товаров в розничной сети по магазинам

**OLAP-куб**

Оси:
* 0: ID группы товаров (100)
* 1: ID магазина (10)
* 2: день месяца, в котором осуществляются продажи (31)

```python
olap = np.zeros((100,10,31))
```

```python
for i in range (olap.shape[0]):
    a = np.random.randint(1,101)
    for j in range (olap.shape[1]):
        b = a + a * np.random.rand()
        for k in range (olap.shape[2]):
            olap[i][j][k] = b * np.random.randint(1,10)
            
```

## III. Для получившегося OLAP-куба посчитать суммарные, средние, минимальные и максимальные продажи (на самом деле любые агрегатные функции можно использовать), зафиксировав значения по одной, двум или трём осям

**Срез для 0-товара по всем дням и магазинам**

```python
olap[0,:,:]
```

```python
print("Максимум= " + str(olap[0,:,:].max()))
print("Минимум= " + str(olap[0,:,:].min()))
print("Среднее= " + str(olap[0,:,:].mean()))
```

**Максимумы продаж по дням 0-товара для всех магазаинов**

```python
print("Максимумы = " + str(olap[0,:,:].max(axis = 1)))
```

**Среднемесячные продажи 10-товара для всех магазаинов**

```python
print("Среднее = " + str(olap[10,:,:].mean(axis = 1)))
```

**Cумма, максимум, минимум и среднее по продажам 10-20 чисел месяца для товаров [1,3,4] по всем магазинам**

```python
print("Сумма= " + str(olap[[1,3,4],:,10:21].sum()))
print("Максимум= " + str(olap[[1,3,4],:,10:21].max()))
print("Минимум= " + str(olap[[1,3,4],:,10:21].min()))
print("Среднее= " + str(olap[[1,3,4],:,10:21].mean()))
```

**Cумма, максимум, минимум и среднее по продажам 0-15 чисел месяца для товаров 26-53 по магазинам 2-7**

```python
print("Сумма= " + str(olap[26:54,2:8,:16].sum()))
print("Максимум= " + str(olap[26:54,2:8,:16].max()))
print("Минимум= " + str(olap[26:54,2:8,:16].min()))
print("Среднее= " + str(olap[26:54,2:8,:16].mean()))
```
